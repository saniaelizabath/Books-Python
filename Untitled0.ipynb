```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Datasets/main/titanic-training-data.csv")
```

```python
df.head()
```

```python

sns.relplot(x="Age", y="Fare", data = df)
plt.show()
```

```python
sns.relplot(x="Age", y="Fare", hue = "Pclass", data = df)
```

```python
sns.relplot(x="Age", y="Fare", size = "Survived", data = df);
```

```python
sns.catplot(x="Sex", y="Age", data=df)
```

```python
sns.catplot(x="Age", y="Pclass", jitter = False,  data=df)
```

```python
sns.catplot(x="Fare", y="Age", hue = "Sex", data=df)
```
